```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def dx(x,t):
    dx0=x[1]
    dx1=-k/m*x[3]
    dx2=x[3]
    dx3=k/m*x[1]-g
    return [dx0,dx1,dx2,dx3]
```

```python
def dx(x,t):
    v=(x[1]**2+x[3]**2)**(0.5)
    theta=np.arctan(x[3]/x[1])
    print(theta)
    Fl=2*np.pi*w0*R*R*L*rho*v
    Fd=0
    dx0=x[1]
    dx1=(-Fl*np.sin(theta)-Fd*np.cos(theta))/m
    dx2=x[3]
    dx3=(-m*g+Fl*np.cos(theta)-Fd*np.sin(theta))/m
    return [dx0,dx1,dx2,dx3]
```

```python
g=9.8 #m/s^2
m=0.02 #kg
w0=36 #rad/sec
R=0.05 #m
L=0.15 #m
rho=1.29 #kg/m^3
k=2*np.pi*w0*(R**2)*L*rho
t=np.linspace(0,10,100)
x0=[0,5,2,0.1]
x=odeint(dx,x0,t)
```

```python
plt.plot(x[:,0],x[:,2],"-.")
```
