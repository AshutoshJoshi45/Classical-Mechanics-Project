```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def dx(x,t):
    dx0=x[1]
    dx1=(-k1*x[3]-k2*x[1]*np.sqrt(x[1]**2+x[3]**2))/m
    dx2=x[3]
    dx3=(-m*g+k1*x[1]-k2*np.sqrt(x[1]**2+x[3]**2)*x[3])/m
    return [dx0,dx1,dx2,dx3]
```

```python
g=9.8 #m/s^2
m=0.05 #kg
w0=80 #rad/sec
R=0.05 #m
L=0.15 #m
rho=1.29 #kg/m^3
A=2*np.pi*R*L
Cd=1 #drag coefficient
k1=2*np.pi*w0*R*R*L*rho
k2=1/2*Cd*A*rho
```

```python
t=np.linspace(0,2,100)
x0=[0,10,0,5]
x=odeint(dx,x0,t)
```

```python
fig=plt.figure()
ax=fig.add_subplot(3,3,1)
plt.plot(x[:,0],x[:,2],".k")
plt.grid()
plt.xlabel("x(m)")
plt.ylabel("y(m)")
ax=fig.add_subplot(3,3,2)
plt.plot(t,x[:,1],".k")
plt.xlabel("t(s)")
plt.ylabel("vx(m/s)")
plt.grid()
ax=fig.add_subplot(3,3,3)
plt.plot(t,x[:,3],".k")
plt.xlabel("t(s)")
plt.ylabel("vy(m/s)")
plt.grid()
fig.suptitle("For Light Mass")
plt.savefig("light_mass.png")
```
